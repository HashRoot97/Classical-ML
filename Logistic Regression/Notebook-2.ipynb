```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
def sigmoid(x):
    return 1/(1+np.exp(-x))
```

```python
def dsigmoid(x):
    return sigmoid(x)*(1-sigmoid(x))
```

```python
dsigmoid(0)
```

```python
x = np.arange(-5, 5, 0.01)
```

```python
sig = sigmoid(x)
plt.plot(x, sig, linewidth=2)
plt.plot(x, dsigmoid(x), linewidth=3)
plt.legend(['sigmoid', 'd(sigmoid)'])
plt.show()
```
